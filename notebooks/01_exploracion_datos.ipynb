```python
matches_without_score = results[
    results["home_score"].isna() | results["away_score"].isna()
]

print("Partidos sin marcador:", matches_without_score.shape)
display(matches_without_score.head(20))
```

```python
worldcup_2026_matches = results[
    (results["tournament"] == "FIFA World Cup") &
    (results["home_score"].isna()) &
    (results["away_score"].isna())
].copy()

print("Partidos Mundial 2026 sin marcador:", worldcup_2026_matches.shape)
display(worldcup_2026_matches.head(30))
```

```python
historical_matches = results.dropna(subset=["home_score", "away_score"]).copy()

def get_result(row):
    if row["home_score"] > row["away_score"]:
        return "home_win"
    elif row["home_score"] < row["away_score"]:
        return "away_win"
    else:
        return "draw"

historical_matches["result"] = historical_matches.apply(get_result, axis=1)

historical_matches["result"].value_counts()
```

```python
historical_matches["result"].value_counts(normalize=True).round(4) * 100
```

```python
historical_matches["year"] = historical_matches["date"].dt.year

matches_by_year = historical_matches.groupby("year").size().reset_index(name="matches")

display(matches_by_year.tail(20))
```
