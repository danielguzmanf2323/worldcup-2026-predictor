```python
from pyspark.sql.functions import current_timestamp, col

base_path = "/Volumes/workspace/default/worldcup_2026"

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze")

results_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/results.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", col("_metadata.file_path"))
)

shootouts_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/shootouts.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", col("_metadata.file_path"))
)

goalscorers_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/goalscorers.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", col("_metadata.file_path"))
)

former_names_raw = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(f"{base_path}/former_names.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", col("_metadata.file_path"))
)

results_raw.write.mode("overwrite").format("delta").saveAsTable("workspace.bronze.results_raw")
shootouts_raw.write.mode("overwrite").format("delta").saveAsTable("workspace.bronze.shootouts_raw")
goalscorers_raw.write.mode("overwrite").format("delta").saveAsTable("workspace.bronze.goalscorers_raw")
former_names_raw.write.mode("overwrite").format("delta").saveAsTable("workspace.bronze.former_names_raw")

print("Tablas Bronze creadas correctamente.")
```
